```python
# SIMULASI SPARK DENGAN PANDAS - PASTI BERHASIL
import pandas as pd
from collections import Counter

print("=== SIMULASI SPARK WORD COUNT ===")

# Data sample
data = ["Hello Spark", "Hello World", "Spark is awesome", "Big Data praktikum"]

# Simulasi RDD - Word Count
print("\n1. SIMULASI RDD:")
all_words = []
for text in data:
    words = text.split(" ")
    all_words.extend(words)

word_count = Counter(all_words)
for word, count in word_count.items():
    print(f"'{word}': {count}")

# Simulasi DataFrame
print("\n2. SIMULASI DATAFRAME:")
df = pd.DataFrame(data, columns=['text'])
df['words'] = df['text'].str.split()
word_df = df.explode('words')
result_df = word_df.groupby('words').size().reset_index(name='count')
result_df = result_df.sort_values('count', ascending=False)

print("Hasil DataFrame-style:")
print(result_df)

print("\n✅ PRAKTIKUM SPARK SIMULASI SELESAI!")
```
