```python
# DATA PREPROCESSING DI COLAB
!pip install pyspark

import pandas as pd
import numpy as np
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

# Data sample dari modul
data_kotor = [
    (1, 'Budi Susanto', 25, 5500000, 'L', '2022-01-15', 'Jakarta', 'Transaksi berhasil'),
    (2, 'Ani Lestari', None, 8000000, 'P', '2022-02-20', 'Bandung', 'Pengiriman cepat'),
    (3, 'Candra Wijaya', 35, 12000000, 'L', '2022-01-18', 'Surabaya', 'Sangat puas'),
    (4, 'Dewi Anggraini', 22, 4800000, 'P', '2022-03-10', 'JKT', 'Barang diterima'),
    (5, 'Budi Susanto', 25, 5500000, 'L', '2022-01-15', 'Jakarta', 'Transaksi berhasil')  # duplikat
]

# Convert ke Pandas DataFrame dulu
df = pd.DataFrame(data_kotor, columns=['id', 'nama', 'usia', 'gaji', 'jk', 'tgl', 'kota', 'ulasan'])
print("DATA AWAL:")
print(df)

# Data Cleaning
print("\n=== DATA CLEANING ===")
# Handle missing values
df['usia'] = df['usia'].fillna(df['usia'].mean())
df['ulasan'] = df['ulasan'].fillna('Tidak ada ulasan')

# Hapus duplikat
df = df.drop_duplicates()

# Standardisasi kota
df['kota'] = df['kota'].replace('JKT', 'Jakarta')

print("DATA SETELAH CLEANING:")
print(df)
```
